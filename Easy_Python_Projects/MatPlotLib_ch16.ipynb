# Looking at daily high and low temps. in Sitka, Alaska in July 2014

```python
import csv
from datetime import datetime
from matplotlib import pyplot as plt

# Get dates, high, and low temperatures from file.
filename = 'sitka_weather_07_2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    dates, highs, lows = [], [], []
    for row in reader:
        try:
            current_date = datetime.strptime(row[0], "%Y-%m-%d")
            high = int(row[1])
            low = int(row[3])
        except ValueError:
            print(current_date, 'missing data')
        else:
            dates.append(current_date)
            highs.append(high)
            lows.append(low)

# Plot data.
fig = plt.figure(dpi=128, figsize=(10, 6))
plt.plot(dates, highs, c='red', alpha=0.5)
plt.plot(dates, lows, c='blue', alpha=0.5)
plt.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

# Format plot.
title = "Daily high and low temperatures - July 2014\nSitka, AK"
plt.title(title, fontsize=20)
plt.xlabel('', fontsize=16)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

# Comparison of daily high temp. in 2014 between Death Valley and Sitka

```python
import csv
from datetime import datetime
from matplotlib import pyplot as plt

# Get dates, high, and low temperatures from file.
filename = 'death_valley_2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    dates_dv, highs_dv, lows_dv = [], [], []
    for row in reader:
        try:
            current_date = datetime.strptime(row[0], "%Y-%m-%d")
            high = int(row[1])
            low = int(row[3])
        except ValueError:
            print(current_date, 'missing data')
        else:
            dates_dv.append(current_date)
            highs_dv.append(high)
            lows_dv.append(low)
            
            
# Get dates, high, and low temperatures from file.
filename = 'sitka_2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    dates_sitka, highs_sitka, lows_sitka = [], [], []
    for row in reader:
        try:
            current_date = datetime.strptime(row[0], "%Y-%m-%d")
            high = int(row[1])
            low = int(row[3])
        except ValueError:
            print(current_date, 'missing data')
        else:
            if current_date not in dates_dv:
                continue
            else:
                dates_sitka.append(current_date)
                highs_sitka.append(high)
                lows_sitka.append(low)
            

# Plot data.
fig = plt.figure(dpi=128, figsize=(10, 6))
plt.plot(dates_sitka, highs_sitka, c='red', alpha=0.5)
plt.plot(dates_dv, highs_dv, c='blue', alpha=0.5)

# Format plot.
title = "Comparison of Daily high temperatures in 2014 between \nDeath Valley, CA and Sitka, AK"
plt.title(title, fontsize=20)
plt.xlabel('', fontsize=16)
#fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

# World population in 2010 by country

```python
from pygal.maps.world import COUNTRIES
import json
import pygal
from pygal.style import LightColorizedStyle as LCS, RotateStyle as RS
from pygal.maps.world import World

def get_country_code(country_name):
    """Return the Pygal 2-digit country code for the given country."""
    for code, name in COUNTRIES.items():
        if name == country_name:
            return code
    # If the country wasn't found, return None.                                                                                                               
    return None


# Load the data into a list.                                                                                                                                  
filename = 'population_data.json'
with open(filename) as f:
    pop_data = json.load(f)

# Build a dictionary of population data.                                                                                                                      
cc_populations = {}
for pop_dict in pop_data:
    if pop_dict['Year'] == '2010':
        country_name = pop_dict['Country Name']
        population = int(float(pop_dict['Value']))
        code = get_country_code(country_name)
        if code:
            cc_populations[code] = population

wm_style = RS('#336699', base_style=LCS)                                                                                                                     
wm = World()
wm.title = 'World Population in 2010, by Country'
wm.add('2010',cc_populations)

wm.render_to_file('world_population.svg')
```

# World population in 2010, now grouped in three categories

```python
import json

import pygal
from pygal.style import LightColorizedStyle as LCS, RotateStyle as RS
from pygal.maps.world import World
from country_codes import get_country_code

# Load the data into a list.                                                                                                                                  
filename = 'population_data.json'
with open(filename) as f:
    pop_data = json.load(f)

# Build a dictionary of population data.                                                                                                                      
cc_populations = {}
for pop_dict in pop_data:
    if pop_dict['Year'] == '2010':
        country_name = pop_dict['Country Name']
        population = int(float(pop_dict['Value']))
        code = get_country_code(country_name)
        if code:
            cc_populations[code] = population

# Group the countries into 3 population levels.                                                                                                               
cc_pops_1, cc_pops_2, cc_pops_3 = {}, {}, {}
for cc, pop in cc_populations.items():
    if pop < 10000000:
        cc_pops_1[cc] = pop
    elif pop < 1000000000:
        cc_pops_2[cc] = pop
    else:
        cc_pops_3[cc] = pop

# See how many countries are in each level.                                                                                                                   
print(len(cc_pops_1), len(cc_pops_2), len(cc_pops_3))

wm_style = RS('#336699', base_style=LCS)
wm = World()
wm.title = 'World Population in 2010, by Country'
wm.add('0-10m', cc_pops_1)
wm.add('10m-1bn', cc_pops_2)
wm.add('>1bn', cc_pops_3)

wm.render_to_file('world_population.svg')
```

# Percentage of forest land in each country in 2020

```python
import csv

import pygal
from pygal.style import LightColorizedStyle as LCS, RotateStyle as RS
from pygal.maps.world import World
from country_codes import get_country_code

# Load the data into a list.                                                                                                                                  
filename = 'API_AG.LND.FRST.ZS_DS2_en_csv_v2_5180790.csv'
cc_forest = {}
with open(filename) as f:
    reader = csv.reader(f)
    next(reader)
    next(reader)
    next(reader)
    next(reader)
    header_row = next(reader)
    for row in reader:
        country = row[0]
        code = get_country_code(country)
        if row[64] == "":
            continue
        else:
            percent_forest_area = float(row[64])
        if code and row[64]!= "":
            cc_forest[code] = percent_forest_area
        #print(code, percent_forest_area)                                                                                                                     

# Group the countries into 3 population levels.                                                                                                               
cc_fors_1, cc_fors_2, cc_fors_3 = {}, {}, {}
for cc, pop in cc_forest.items():
    if pop < 15.0:
        cc_fors_1[cc] = pop
    elif pop < 33:
        cc_fors_2[cc] = pop
    else:
        cc_fors_3[cc] = pop

# See how many countries are in each level.                                                                                                                   
print(len(cc_fors_1), len(cc_fors_2), len(cc_fors_3))

wm_style = RS('#336699', base_style=LCS)
wm = World(style = wm_style)
wm.title = 'World Forest Area coverage,2020 by Country'
wm.add('0-15%', cc_fors_1)
wm.add('15-33%', cc_fors_2)
wm.add('>33%', cc_fors_3)

wm.render_to_file('world_forest.svg')
```
